# Q1. Which nominee has received the most awards overall or in a single year?

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['win'].unique())
df['win'] = le.transform(df['win'])
df.head()
```

```python
df.tail()
```

```python
total_nominee = df.pivot_table('win', index='year_award', columns='nominee', aggfunc=sum)
total_nominee
```

```python
getscore = pd.DataFrame(total_nominee.sum(axis=0))
getscore.columns = ['counts']
getscore.sort_values(by='counts', ascending=False)[:10]
```
